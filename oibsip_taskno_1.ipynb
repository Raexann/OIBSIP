Loading required libararies

```python
import pandas as pd
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Loading dataset from sklearn

```python
data=datasets.load_iris()
```

```python
data
```

```python
data.keys()
```

Creating dataframe from the given data

```python
iris=pd.DataFrame(data=np.c_[data['data'],data['target']],columns=data['feature_names']+['target'])
```

```python
iris
```

Converting target column into categorical

```python
species=[]
for i in range(len(iris['target'])):
    if iris['target'][i]==0:
        species.append('Setosa')
    elif iris['target'][i]==1:
        species.append('Versicolor')
    else:
        species.append('Virginica')
```

```python
iris['species']=species
```

Exploratory Data Analysis

```python
species_values=iris.groupby('species').size().values
Species=['Setosa','Versicolor','Virginica']
plt.bar(range(len(species_values)),species_values,label=('Setosa','Versicolor','Virginica'),color=('orange','red', 'blue'))
plt.xticks([0.3,1.3, 2.3],Species)
plt.ylabel("No of people Survived")
plt.xlabel("Species")
plt.title("Count")
plt.legend(loc='upper right')
plt.show()
```

```python
sns.heatmap(iris.corr(),annot=True,vmin=-0.5,vmax=0.6)
```

```python
sns.pairplot(iris, x_vars=iris.iloc[:,0:4], y_vars=iris.iloc[:,0:4], hue='species')
```

Training and building the model

```python
X=iris.drop(['species','target'], axis=1)
y=iris['species']
```

```python
X
```

```python
y
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,
                                               random_state=101)
```

```python
print(X_train.shape, X_test.shape)
print(Y_train.shape, Y_test.shape)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
prediction=model.predict(X_test)
```

```python
prediction
```

Model Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print("Classification Report")
print(classification_report(Y_test,prediction))
print("Confusion Matrix")
print(confusion_matrix(Y_test,prediction))
```

Accuracy

```python
accuracy=accuracy_score(Y_test, prediction)*100
```

```python
print('Accuracy of the model is {:.2f}'.format(accuracy))
```
